```python
# import library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# load data
customer_lifespan_data = pd.read_csv('https://docs.google.com/spreadsheets/d/1by8tPHwOnq3uKYK2E7sA9VBUYoPM4p1Rnrm_Ss9cyHI/export?format=csv')
daily_mkt_spendings = pd.read_csv('https://docs.google.com/spreadsheets/d/1AZOIThOV4P-0eYDge53ZwumVkfkHoYPWxst3k3Bv87c/export?format=csv')
monthly_expenses = pd.read_csv('https://docs.google.com/spreadsheets/d/10OGbaywwMIqKgnPGy8VDvpBVtjyqln47iYa2lFhI9Mw/export?format=csv')
payroll = pd.read_csv('https://docs.google.com/spreadsheets/d/1c_WihqTZCQvNgxzmd-OwhR9i5diwtfxXVLyMn8R-Lp4/export?format=csv')
receipts_history = pd.read_csv('https://docs.google.com/spreadsheets/d/1qayqML1zCKdmtzutkcy9LWvE6xFRm6TGBEVkHHJKIuE/export?format=csv')
```

```python
# Total spending includes: system, salary & marketing expenses
spending_system = monthly_expenses[(monthly_expenses['month']=='2023-03') & (monthly_expenses['item']=='Salesforce')]['amount'].sum()
salary = payroll[(payroll['month']=='2023-03') & (payroll['department'].isin(['Sales', 'Marketing']))]['paid'].sum()
mkt_expense = daily_mkt_spendings[(daily_mkt_spendings['date'] >= '2023-03-01') & (daily_mkt_spendings['date'] <= '2023-03-31')]['spending'].sum()
total_spending = spending_system + salary + mkt_expense

# New CX in March 2023
mar_new_customer = receipts_history[(receipts_history['date'] >= '2023-03-01') & (receipts_history['date'] <= '2023-03-31')
                                  & (receipts_history['new_customer'] == 1)]['customer_id'].nunique()

# CAC
CAC = total_spending/mar_new_customer
print(f'CAC: {CAC}')
```

```python
# ARPU
mar_total_revenue = receipts_history[(receipts_history['date'] >= '2023-03-01' ) & (receipts_history['date'] <= '2023-03-31')]['receipt_amount'].sum()
mar_total_cx = receipts_history[(receipts_history['date'] >= '2023-03-01' ) & (receipts_history['date'] <= '2023-03-31')]['customer_id'].nunique()

ARPU = mar_total_revenue / mar_total_cx
print(f'ARPU: {ARPU}')
```

```python
# COG
service_fee = monthly_expenses[monthly_expenses['month'] == '2023-03']['amount'].sum()
salary = payroll[payroll['month'] == '2023-03']['paid'].sum()

COG = service_fee*0.6 + salary
print(f'COG: {COG}')
```

```python
# Gross Margin
profit = mar_total_revenue - COG
gross_margin = profit / mar_total_revenue * 100
print(f'Gross Margin: {gross_margin}')
```

```python
# LTV
customer_lifespan = ( pd.to_datetime(customer_lifespan_data['churn_date']) - pd.to_datetime(customer_lifespan_data['start_date']) ).dt.days.mean()/30
LTV = ARPU * customer_lifespan * gross_margin
print(f'LTV: {LTV}')
```

```python
# LTV / CAC
LTV_CAC = LTV / CAC
print(f'LTV / CAC: {LTV_CAC}')
```

```python
# Summary
print(f'CAC: {CAC}')
print(f'ARPU: {ARPU}')
print(f'COG: {COG}')
print(f'Gross Margin: {gross_margin}')
print(f'LTV: {LTV}')
print(f'LTV / CAC: {LTV_CAC}')
```
